```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bissecção(F,A,B,TOL,PARAM):
    E = (B-A)/2
    p = (A+B)/2
    
    while E > TOL:
        if F(A,PARAM)*F(p,PARAM)<0:
            B=p
        else:
            A=p
        E = (B-A)/2
        p = (A+B)/2
    
    return p, E
```

```python
def eqdipolo(r,parametros):
    angulo = parametros[0]
    v0 = parametros[1]
    
    vr0 = 1/np.sqrt(r**2 + 2*r*np.cos(angulo) + 1) - 1/r
    
    return vr0 - v0
```

```python
Na = 100
angulos = np.linspace(0, 2*np.pi, Na)
raios = np.zeros(Na)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) 
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

plt.plot(0, 0,"bo")
plt.plot(-1, 0,"ro")
plt.axvline(x=-0.5,color = 'black')

for v0 in [-3,-2,-1,-0.5,-0.25,-0.1]:
    for i in range(Na):
        p,E = bissecção(eqdipolo,1e-3,20,1e-5,[angulos[i], v0])
        raios[i] = p
    
    x = raios*np.cos(angulos)
    y = raios*np.sin(angulos)
    
    plt.plot(x,y,"b-")
    plt.plot(-x-1,y,"r-")
```
